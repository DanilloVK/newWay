## Opdracht 1: one hot encoding

**Schrijf code die deze lijst omzet in one-hot vectors. Je mag geen library gebruiken voor de one hot encoder**

`categories = ["red", "blue", "green", "blue", "red"]`

Hints:

- Maak een lijst met unieke waarden

- Maak een vector van 0’en voor elke categorie

- Zet een 1 op de juiste positie

```python
categories = ["red", "blue", "green", "blue", "red"]

# Stap 1: verzamel alle unieke waarden in een lijst
unieke_waarden = []
for categorie in categories:
    if categorie not in unieke_waarden:
        unieke_waarden.append(categorie)

print("Unieke waarden:", unieke_waarden)

# Stap 2: maak per categorie een one-hot vector
one_hot_vectoren = []
for categorie in categories:
    # Begin met een vector van alleen nullen
    vector = [0] * len(unieke_waarden)

    # Zoek de positie van deze categorie in de unieke-lijst
    positie = unieke_waarden.index(categorie)

    # Zet een 1 op die positie
    vector[positie] = 1

    one_hot_vectoren.append(vector)

# Resultaat tonen
print(f"\nKolommen: {unieke_waarden}")
for categorie, vector in zip(categories, one_hot_vectoren):
    print(f"  {categorie:>6} -> {vector}")
```

### Opdracht 2 Datastructuren met one hot encoding
Nu gaan we er van uit dat de lijst groter is

```python
import random

colors = ["red", "blue", "green", "yellow", "purple"]
categories = [random.choice(colors) for _ in range(20000)]
```

```python
import time

# Methode 1: LIST met .index()
# Dezelfde aanpak als Opdracht 1, maar nu op 20.000 items

unieke_lijst = []
for categorie in categories:
    if categorie not in unieke_lijst:
        unieke_lijst.append(categorie)

start = time.time()

one_hot_list_resultaat = []
for categorie in categories:
    vector = [0] * len(unieke_lijst)
    positie = unieke_lijst.index(categorie)   # .index() doorzoekt de hele lijst -> O(n)
    vector[positie] = 1
    one_hot_list_resultaat.append(vector)

tijd_list = time.time() - start
print(f"Methode LIST (.index): {tijd_list:.4f} seconden")
print(f"Aantal vectoren: {len(one_hot_list_resultaat)}")
print(f"Voorbeeld: {categories[0]} -> {one_hot_list_resultaat[0]}")
```

Probeer het ook met een dictionary.


Maak een vergelijking op tijd voor al de 3 methodes. Wat valt je op? Bespreek.

```python
# Methode 2: DICTIONARY
# Een dict slaat de positie per categorie op -> opzoeken is O(1)

categorie_naar_index = {}
teller = 0
for categorie in categories:
    if categorie not in categorie_naar_index:
        categorie_naar_index[categorie] = teller
        teller += 1

print("Mapping:", categorie_naar_index)

start = time.time()

one_hot_dict_resultaat = []
for categorie in categories:
    vector = [0] * len(categorie_naar_index)
    positie = categorie_naar_index[categorie]  # O(1) dankzij hashing
    vector[positie] = 1
    one_hot_dict_resultaat.append(vector)

tijd_dict = time.time() - start
print(f"Methode DICT: {tijd_dict:.4f} seconden")
print(f"Voorbeeld: {categories[0]} -> {one_hot_dict_resultaat[0]}")
```

Probeer je eerder gemaakte code nu te runnen. Wat valt je op? Waarom is .index zo traag

```python
# .index() is traag omdat het de HELE lijst doorloopt tot het een match vindt.
# Bij elke aanroep begint het weer bij het begin -> O(n) per keer.
# Voor 20.000 items en 5 unieke waarden: 20.000 x gemiddeld 2.5 vergelijkingen.
#
# Een dictionary daarentegen gebruikt HASHING:
# het berekent een hash van de key en springt direct naar de juiste plek -> O(1).
#
# Verschil: list.index() = lineair zoeken, dict[key] = hash lookup.

print(f"Ter vergelijking:")
print(f"  LIST (.index):  {tijd_list:.4f} sec")
print(f"  DICT (hashing): {tijd_dict:.4f} sec")
print(f"  Dict is ~{tijd_list / tijd_dict:.1f}x sneller")
```

**Probeer nu helzelfde te implementeren met een SET**

```python
# Methode 3: SET
# Een set vindt unieke waarden razendsnel, maar heeft GEEN index/volgorde.
# We moeten de set omzetten naar een lijst om posities te kunnen gebruiken.

unieke_set = set(categories)
unieke_als_lijst = list(unieke_set)

print("Unieke waarden (via set):", unieke_als_lijst)

start = time.time()

one_hot_set_resultaat = []
for categorie in categories:
    vector = [0] * len(unieke_als_lijst)
    positie = unieke_als_lijst.index(categorie)  # Nog steeds .index() op een lijst!
    vector[positie] = 1
    one_hot_set_resultaat.append(vector)

tijd_set = time.time() - start
print(f"Methode SET: {tijd_set:.4f} seconden")

# De set helpt bij het snel vinden van unieke waarden,
# maar het encoden zelf is even traag als de list-methode
# want we gebruiken weer .index() op een lijst.

print(f"\n{'='*40}")
print(f"VERGELIJKING:")
print(f"  LIST (.index):  {tijd_list:.6f} sec")
print(f"  DICT (hashing): {tijd_dict:.6f} sec")
print(f"  SET + .index(): {tijd_set:.6f} sec")
print(f"{'='*40}")
print(f"Dictionary is veruit het snelst voor encoding!")
```

**Vragen:**

- Welke datastructuur was het snelst?

- Wat maakt een dictionary bijzonder t.o.v. een list?

- Waarom is .index() op een list traag?

- Waarom helpt een set alleen bij het vinden van unieke elementen, maar niet bij het encoden?

- Wat zou er gebeuren met miljoenen elementen? Probeer het!

- Probeer het te runnen met een ander aantal unieke labels. Wat kun je hier over zeggen?

- Welke datastructuur zou je kiezen in real-world machine learning preprocessing?

**Antwoorden:**

- **Welke datastructuur was het snelst?**
  De **dictionary** is veruit het snelst voor one-hot encoding. Opzoeken kost O(1) door hashing, terwijl `.index()` op een list O(n) kost.

- **Wat maakt een dictionary bijzonder t.o.v. een list?**
  Een dictionary gebruikt **hashing** om keys direct te koppelen aan waarden. Hierdoor is opzoeken O(1), terwijl een list lineair moet zoeken O(n).

- **Waarom is `.index()` op een list traag?**
  `.index()` begint bij het eerste element en loopt de hele lijst door tot het een match vindt. Bij elk element wordt een vergelijking gedaan -> O(n) per aanroep.

- **Waarom helpt een set alleen bij het vinden van unieke elementen, maar niet bij het encoden?**
  Een set is snel in het checken of een element erin zit (O(1)), maar een set heeft **geen volgorde en geen index**. Om te encoden heb je een positie nodig, dus moet je de set alsnog omzetten naar een lijst en `.index()` gebruiken.

- **Wat zou er gebeuren met miljoenen elementen?**
  Met miljoenen elementen wordt het verschil nog groter. De list-methode wordt extreem traag (minuten), terwijl de dict-methode nauwelijks langzamer wordt. De tijdscomplexiteit bepaalt de schaalbaarheid.

- **Welke datastructuur zou je kiezen in real-world ML preprocessing?**
  Een **dictionary** voor de mapping van categorie naar index. In de praktijk gebruiken libraries zoals scikit-learn ook intern dictionaries/hashmaps voor dit doel.

## Opdracht 3: Ordinal Encoding



```python

# Encodeer onderstaande reviews als numerieke waarden
# Bepaal zelf de beste volgorde

reviews = ["goed", "matig", "matig", "slecht", "uitstekend"]
```

```python
# Ordinal encoding: labels met een logische RANGORDE
# krijgen een oplopende numerieke waarde.
# De volgorde is belangrijk: slecht < matig < goed < uitstekend

volgorde = {
    "slecht":      0,
    "matig":       1,
    "goed":        2,
    "uitstekend":  3,
}

encoded_reviews = []
for review in reviews:
    encoded_reviews.append(volgorde[review])

for review, waarde in zip(reviews, encoded_reviews):
    print(f"  {review:>12} -> {waarde}")

print(f"\nOrigineel:  {reviews}")
print(f"Encoded:    {encoded_reviews}")
```

## Opdracht 4: Binning toepassen

Maak drie categorieën (Jong, Midden, Oud):

- Bin 1: laagste derde
- Bin 2: middelste derde
- Bin 3: hoogste derde


Maak een functie die leeftijden in 3 bins stopt op basis van hun waarde.

Gebruik equal-frequency.


```python
leeftijden = [18, 22, 25, 30, 35, 40, 45, 50, 55, 60, 65, 70]

def equal_frequency_binning(data, aantal_bins=3):
    """Verdeelt data in bins met (ongeveer) evenveel elementen per bin."""
    gesorteerd = sorted(data)
    grootte_per_bin = len(gesorteerd) // aantal_bins

    bins = []
    for i in range(aantal_bins):
        start_index = i * grootte_per_bin

        if i == aantal_bins - 1:
            # Laatste bin pakt alles wat overblijft
            bins.append(gesorteerd[start_index:])
        else:
            bins.append(gesorteerd[start_index : start_index + grootte_per_bin])

    return bins

labels = ["Jong", "Midden", "Oud"]
bins = equal_frequency_binning(leeftijden, aantal_bins=3)

print("Bins (equal-frequency):")
for label, bin_inhoud in zip(labels, bins):
    print(f"  {label:>7}: {bin_inhoud}")

# Label toekennen aan elke leeftijd
def geef_label(leeftijd, bins, labels):
    """Zoekt in welke bin een leeftijd valt en geeft het bijbehorende label."""
    for i, bin_inhoud in enumerate(bins):
        if leeftijd in bin_inhoud:
            return labels[i]
    return "Onbekend"

print("\nLabel per leeftijd:")
for leeftijd in leeftijden:
    label = geef_label(leeftijd, bins, labels)
    print(f"  {leeftijd} jaar -> {label}")
```

## Opdracht 5: Normalisatie

Normaliseer handmatig onderstaande getallen zodat ze tussen 0 en 1 liggen.

```python
waarden = [18, 22, 30, 45, 60]
```

```python
# Min-max normalisatie: schaalt elke waarde naar het bereik [0, 1]
# Formule: genormaliseerd = (x - min) / (max - min)

minimum = min(waarden)
maximum = max(waarden)

print(f"Minimum: {minimum}")
print(f"Maximum: {maximum}")
print(f"Bereik:  {maximum - minimum}")
print()

genormaliseerd = []
for waarde in waarden:
    norm = (waarde - minimum) / (maximum - minimum)
    genormaliseerd.append(round(norm, 4))

print("Resultaat:")
for origineel, norm in zip(waarden, genormaliseerd):
    print(f"  {origineel:>3} -> {norm:.4f}")

print(f"\nOrigineel:      {waarden}")
print(f"Genormaliseerd: {genormaliseerd}")
```
