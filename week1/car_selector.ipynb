```python
print("Welkom bij de Car Picker AI")
print("Beantwoord de vragen zodat ik de beste auto voor je kan kiezen.\n")

# input van gebruiker
budget = input("Wat is je budget? (laag / midden / hoog): ").lower()
gezinsgrootte = input("Hoeveel personen vervoer je meestal? (1-2 / 3-5 / 6+): ")
gebruik = input("Waar gebruik je de auto vooral voor? (stad / lange ritten / offroad): ").lower()
milieu = input("Vind je milieuvriendelijkheid belangrijk? (ja / nee): ").lower()

# AI regels
auto = ""

if gezinsgrootte == "6+":
    auto = "Volkswagen Multivan"

elif gebruik == "offroad":
    if budget == "hoog":
        auto = "Land Rover Defender"
    else:
        auto = "Dacia Duster"

elif milieu == "ja":
    if budget == "laag":
        auto = "Renault Zoe"
    elif budget == "midden":
        auto = "Hyundai Kona Electric"
    else:
        auto = "Tesla Model 3"

elif gebruik == "lange ritten":
    if budget == "laag":
        auto = "Skoda Octavia"
    elif budget == "midden":
        auto = "BMW 3 Serie"
    else:
        auto = "Mercedes E-klasse"

elif gebruik == "stad":
    if budget == "laag":
        auto = "Toyota Aygo"
    elif budget == "midden":
        auto = "Volkswagen Polo"
    else:
        auto = "Mini Cooper"

else:
    auto = "Toyota Corolla"

print("\nOp basis van jouw antwoorden adviseer ik:")
print("ðŸ‘‰", auto)
```
