```python
# Bereken de gemiddelde 
mean_image = np.mean(X_train, axis=0)
plt.imshow(mean_image, cmap='gray')
plt.title('Gemiddelde afbeelding')
plt.axis('off')
plt.show()

# Bereken de gemiddelde waarde van standaardafwijking van de pixels voor alle cijfers inne de training set
std_image = np.std(X_train, axis=0)
plt.imshow(std_image, cmap='gray')
plt.title('Standaardafwijking van pixels')
plt.axis('off')
plt.show()

# print het totale gemiddelde / standaardafwijkingen en het gemiddelde stndaardafwijking per cijfer
overall_mean = np.mean(X_train)
overall_std = np.std(X_train)
print(f'Totale gemiddelde pixelwaarde: {overall_mean:.4f}')
print(f'Totale standaardafwijking van pixelwaarden: {overall_std:.4f}') 

# Wat zou je kunnen met de stdev en de gemiddelde waarden
```


De totale gemiddelde pixelwaarde geeft aan hoe helder de afbeeldingen gemiddeld zijn. <br>
Een waarde dicht bij 0 betekent dat de afbeeldingen over het algemeen donker zijn, terwijl een waarde dicht bij 1 betekent dat de afbeeldingen over het algemeen helder zijn. <br>
De totale standaardafwijking van pixelwaarden geeft aan hoe variabel de pixelwaarden zijn in de dataset. <br>
Een hoge standaardafwijking betekent dat er veel variatie is in de helderheid van de afbeeldingen, terwijl een lage standaardafwijking betekent dat de afbeeldingen over het <br>algemeen vergelijkbaar zijn in helderheid.

De gemiddelde pixelwaarde en standaardafwijking kunnen gebruikt worden om de distributie van pixelwaarden in de dataset te analyseren.
Dit helpt bij het begrijpen van de variabiliteit in helderheid binnen de dataset, wat belangrijk is voor het optimaliseren van modellen.
Bijvoorbeeld: als de standaardafwijking hoog is, betekent dit dat er veel variatie is in helderheid, wat kan wijzen op een rijke dataset of een complexe taak.
